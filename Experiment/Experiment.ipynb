```python
import pandas as pd
import numpy as np
import mysql.connector
import ChildDFs
import clusterSets
import ExecuteAllocation
import ClusterMetrics
import RouteGeneration
import allo1
import allo2
import randomAllo
```

```python
def randomGeneration(resultDF):
    algoRandomDF = pd.DataFrame()
    algo2DF = pd.DataFrame()
    algo3DF = pd.DataFrame()
    algo6DF = pd.DataFrame()
    algo9DF = pd.DataFrame()
    algo12DF = pd.DataFrame()
    algo15DF = pd.DataFrame()
    algo18DF = pd.DataFrame()
    for i in range(0, 10):
        randomChild = ChildDFs.childDFRandomSort()
        newRowRandom = randomAlgo()
        algoRandomDF = pd.concat([algoRandomDF, newRowRandom], ignore_index=True)
        newRow2 = algo2(randomChild)
        algo2DF = pd.concat([algo2DF, newRow2], ignore_index=True)
        newRow3 = algo3(randomChild)
        algo3DF = pd.concat([algo3DF, newRow3], ignore_index=True)
        newRow6 = algo6(randomChild)
        algo6DF = pd.concat([algo6DF, newRow6], ignore_index=True)
        newRow9 = algo9(randomChild)
        algo9DF = pd.concat([algo9DF, newRow9], ignore_index=True)
        newRow12 = algo12(randomChild)
        algo12DF = pd.concat([algo12DF, newRow12], ignore_index=True)
        newRow15 = algo15(randomChild)
        algo15DF = pd.concat([algo15DF, newRow15], ignore_index=True)
        newRow18 = algo18(randomChild)
        algo18DF = pd.concat([algo18DF, newRow18], ignore_index=True)
        
    df_list = [algoRandomDF, algo2DF, algo3DF, algo6DF, algo9DF, algo12DF, algo15DF, algo18DF]
    
    for df in df_list:
        algorithm_name = df.loc[0, "Algorithm"]
        print(f"Algorithm_Name: {algorithm_name}")
        df.drop(columns=["Algorithm"], inplace=True)
        average_row = df.mean(axis=0)
        average_row["Algorithm"] = algorithm_name
        average_row_df = pd.DataFrame([average_row])
        resultDF = pd.concat([resultDF, average_row_df], ignore_index=True)
        
    print("--------------------------------------------------------------------------------------------------------------------------------------------------")
    print("Collection of Randomness Complete\n\n\n\n\n")
        
    return resultDF
```

```python
def realAllo():
    print("\n\n\n--------------------------------------------------------------------------------------------------------------------------------------------------")
    print("Real Allocation\n\n\n")
    silhouetteE, silhouetteM, db_index, ch_index, wcssE, wcssM, ari, mi, accuracy = ClusterMetrics.runClusterMetrics("locationdatareal")
    min_distance, average_distance, max_distance, var_distance = RouteGeneration.calculateDistances("locationdatareal")
    
    new_row = {
            "Algorithm" : "Real_Allocation",
            "Silhouette_Score_Euclidean" : silhouetteE,
            "Silhouette_Score_Manhattan" : silhouetteM,
            "Davies-Bouldin" : db_index,
            "Calinski-Herabasz" : ch_index,
            "WCSS_Euclidean" : wcssE,
            "WCSS_Manhattan" : wcssM,
            "ARI" : ari,
            "Mutual_Information" : mi,
            "Accuracy" : accuracy,
            "Route_Sim_Min" : min_distance,
            "Route_Sim_Avg" : average_distance,
            "Route_Sim_Max" : max_distance,
            "Route_Sim_Variance" : var_distance
    }
    
    new_row_df = pd.DataFrame([new_row])
    
    return new_row_df
```
